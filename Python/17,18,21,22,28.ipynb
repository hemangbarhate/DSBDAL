```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np 
import seaborn as sns
```

```python
df = pd.read_csv('Admission_Predict.csv')
df
```

# Histogram

```python
df['GRE Score'].plot( kind='hist' , bins=15 )
plt.xlabel('GRE Score')
plt.ylabel('Freq')
plt.title('Distribution of GRE scores')
plt.show()
```

# Dotplot

```python

plt.scatter( df['Chance of Admit '] , df['GRE Score'])
plt.xlabel('Chance of Admit ')
plt.ylabel('GRE Score')
plt.title("Chance of Admit  vs GRE Score")
plt.show()
```

```python
df['University Rating'].value_counts().sort_index().plot(kind='bar')
plt.xlabel('University Rating')
plt.ylabel('Count')
plt.title('Distribution of University Ratings')
plt.show()

```

# Linechart

```python
df.plot(x='Chance of Admit ', y='CGPA')
plt.xlabel('Chance of Admit ')
plt.ylabel('CGPA')
plt.title('Chance of Admit vs. CGPA')
plt.show()
```

```python
plt.scatter( df['GRE Score'], df['TOEFL Score'])
plt.xlabel('GRE Score')
plt.ylabel('TOEFL Score')
plt.title('GRE Score vs. TOEFL Score')
plt.show()
```
